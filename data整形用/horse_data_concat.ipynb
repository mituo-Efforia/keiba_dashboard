# horse_data整形用notebook  
分割して取得したhorsedataを結合して、簡単な整形をしたデータ  

```python
import pandas as pd
```

```python
#分割して取得したhorsedataを読み込む
df_1 =pd.read_csv('../data/horse_data分割/horse_data_201305037.csv',index_col=0)
df_2 = pd.read_csv('../data/horse_data分割/horse_data_201310493.csv',index_col=0)
df_3 = pd.read_csv('../data/horse_data分割/horse_data_2015103689.csv',index_col=0)
df_4 = pd.read_csv('../data/horse_data分割/horse_data_2018106259.csv',index_col=0)
```

```python
#pd.concatで結合
df_list = [df_1,df_2,df_3,df_4]
horse_data_df = pd.concat(df_list)
```

```python
horse_data_df
```

```python
#カラムの半角スペースを削除　
horse_data_df.rename(columns=lambda s: s.replace( ' ',''),inplace=True)
horse_data_df.columns
```

```python
#映像、厩舎コメント、備考、馬場指数、ﾀｲﾑ指数は欠損値しかないので削除
horse_data_df.drop(columns=['映像','厩舎ｺﾒﾝﾄ','備考','馬場指数','ﾀｲﾑ指数'],inplace=True)
horse_data_df
```

```python
#距離と芝を分ける(距離は数字型)
horse_data_df['surface'] = horse_data_df['距離'].str[0]
horse_data_df['距離'] = horse_data_df['距離'].str[1:].astype(int)
```

```python
horse_data_df
```

```python
horse_data_df.to_csv('horse_data_2012_2021.csv',header=True)
```
