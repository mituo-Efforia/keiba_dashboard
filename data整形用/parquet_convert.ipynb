# parquetファイルへの変換  
csvファイルだとhorse_data_2012_2021が253MBもあり、githubにあげられない  
pyarrow(バイナリファイル)に変換すれば圧縮できた

```python
import pandas as pd
import pyarrow
import pyarrow.parquet as pq
```

# horse_data.csvをparquetファイルに変換する

```python
horse_data = pd.read_csv('../data/horse_data_2012_2021.csv',index_col=0)
```

```python
horse_data
```

```python
#parquetファイルへの変換の際にintとstrの両方のデータ型が入っている場合、エラーが出るため、一度str型に変換
#本当はint型で保存したかったが、海外レースは英字が入ってしまうため、str型に変換する。
horse_data['race_id'] = horse_data['race_id'].astype(str)
```

```python
#parquetファイルへの書き出し
horse_data.to_parquet('horse_data_2012_2021.parquet')
```
